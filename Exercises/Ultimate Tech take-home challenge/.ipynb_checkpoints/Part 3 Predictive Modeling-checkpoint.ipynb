```python
import pandas as pd
import pandas_profiling
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
json_file = 'ultimate_data_challenge.json'
json_data = json.load(open(json_file))
df = pd.io.json.json_normalize(json_data)
df.head()
```

```python
df.dtypes
```

```python
df['last_trip_date'] = pd.to_datetime(df['last_trip_date'])
df['signup_date'] = pd.to_datetime(df['signup_date'])
```

```python
df['last_trip_date'].value_counts().sort_index().tail()
```

It looks like this data was pulled on July 1. I will mark everyone whose last trip date is more than 30 days before July 1 as churned.

```python
churn_date = pd.datetime(2014,7,1) - pd.Timedelta('30 days')
df['churned'] = df['last_trip_date'] < churn_date
```

```python
df.profile_report()
```

#### Cleaning
- There are 8 duplicate rows that need to be removed.  
- avg_rating_of_driver has too many missing values to be used  
- phone has 396 missing values. I will remove these rows <1%.  
- The date variables can not be used in modeling.

##### 37.6% of users were retained.

```python
df = df.drop(df[df.duplicated()].index)
df = df.drop(['avg_rating_of_driver', 'last_trip_date', 'signup_date'], axis=1)
df = df.dropna()
```

```python
df.shape
```

```python
f, axes = plt.subplots(3, 3, figsize=(15,8))
for k, col in enumerate(df.loc[:,df.dtypes != 'object'].columns):
    sns.boxplot(x='churned', y=col, ax=axes[k//3, k%3], data=df.loc[:,df.dtypes != 'object'])
```

```python
f, axes = plt.subplots(2, 3, figsize=(15,8))
for k, col in enumerate(df.loc[:,(df.dtypes == 'float64') | (df.dtypes == 'int64')].columns):
    sns.distplot(df.loc[df['churned'], col], ax=axes[k//3, k%3], color='b')
    sns.distplot(df.loc[~df['churned'], col], ax=axes[k//3, k%3], color='orange')
```

```python
f, axes = plt.subplots(1, 3, figsize=(17,5))
sns.countplot(df['city'], ax=axes[0], hue=df['churned'])
sns.countplot(df['phone'], ax=axes[1], hue=df['churned'])
sns.countplot(df['ultimate_black_user'], ax=axes[2], hue=df['churned'])
```

#### Insights
- 3 of the 7 numeric features have a large percentage of zeros.  
- I will replace outlier values in avg_dist, surge_pct, and trips_in_first_30_days with the value at the edge of the whisker.  
- Users from King's Landing are much less likely to have churned than the overall population.  
- Android users are much more likely to churn.  
- Ultimate black users churn at a lower rate.

```python
def remove_outlier(df, col):
    quartiles = np.percentile(df[col].dropna(), [25, 75])
    iqr = quartiles[1] - quartiles[0]
    whisker = quartiles[1] + (iqr * 1.5)
    df.loc[df[col] > whisker, col] = whisker
    
remove_outlier(df, 'avg_dist')
remove_outlier(df, 'surge_pct')
remove_outlier(df, 'trips_in_first_30_days')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegressionCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, auc, roc_curve
```

```python
X = df.iloc[:, :-1].copy()
y = df.iloc[:, -1].copy()
X = pd.get_dummies(X, drop_first=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
lr = LogisticRegressionCV()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
print(classification_report(y_pred, y_test))
```

```python
print(confusion_matrix(y_pred, y_test))
```

```python
probs = lr.predict_proba(X_test)
preds = probs[:,1]
fpr, tpr, threshold = roc_curve(y_test, preds)
roc_auc = auc(fpr, tpr)

# method I: plt
import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print(classification_report(y_pred, y_test))
```

```python
print(confusion_matrix(y_pred, y_test))
```

```python
probs = rf.predict_proba(X_test)
preds = probs[:,1]
fpr, tpr, threshold = roc_curve(y_test, preds)
roc_auc = auc(fpr, tpr)

# method I: plt
import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
pd.DataFrame(rf.feature_importances_, index=X.columns).sort_values(by=0, ascending=False)
```

The 2 most important features in the random forest model are average distance and weekday percent. The 75th percentile of those who stayed with Ultimate is lower than those who left. Customers who churned were more likely to either use Ultimate on only weekdays or to never use it on a weekday.
